```python
import pandas as pd
import numpy as np
import selenium
import time
import tqdm
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

#ignore the all warnings
import warnings
warnings.filterwarnings(action= 'ignore')
option = webdriver.ChromeOptions()
chrome_driver = webdriver.Chrome(options = option)
url = 'https://bina.az/'
chrome_driver.get(url)
```

```python
soup = BeautifulSoup(chrome_driver.page_source, 'html.parser')
```

```python
links = []
link = soup.findAll('a',attrs={"class":"item_link"})
for i in link:
    links.append(' https://bina.az'+i.get('href'))
links
```

```python
locations=[]
location = soup.findAll('div', attrs={"class":"location"})
for i in location:
    locations.append(i.text)
locations
```

```python
dates=[]
date=soup.findAll('div', attrs={"class":"city_when"})
```

```python
dates_times = [date_time.text.split(',')[1].strip() for date_time in soup.find_all('div', class_='city_when')]
```

```python
for i in dates_times:
    dates.append(i)
dates
```

```python
prices=[]
price=soup.findAll('div', attrs={"class":"price"})
for i in price:
    prices.append(i.text)
prices
```

```python
pd.DataFrame(zip(locations,links,dates,prices), columns=['Ərazi','Elan Linki', 'Elan Tarixi', 'Qiymət'])
```
